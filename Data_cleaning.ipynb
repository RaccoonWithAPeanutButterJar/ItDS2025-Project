Importing libraries and loading data

```python
import pandas as pd
import numpy as np

data = pd.read_csv("ai_job_trends_dataset.csv")
```

Checking for empty rows

```python
data.isna().sum()
#Since there are no empty columns, there is no need to fix
```

Looking for incorrect data types

```python
print(data.dtypes)
#It seems that all the data types are already correct and in pandas the object type is the correct data type for text/string columns
```

Checking for duplicates

```python
data.duplicated().sum()
#No duplicates
```

Lookign for outliers

```python
data.describe()
#Taken into account what this dataset is about, there is no possible outliers honestly
```

Checking common rules

```python
data[data['Median Salary (USD)'] <= 0]
data[data['Experience Required (Years)'] < 0]
data[data['Job Openings (2024)'] < 0]
data[data['Projected Openings (2030)'] < 0]
data[data['Remote Work Ratio (%)'] < 0]
data[data['Automation Risk (%)'] < 0]
data[data['Gender Diversity (%)'] < 0]
#No rules broken
```
