## Introduction
This notebook explores the dataset “AI Impact on Job Market (2024–2030)”.
It's purpose is to examine current and projected job trends across industries, analyse job-level characteristics such as automation risk, projected growth, education requirements, and salaries, and highlight patterns that can support deeper analysis in later notebooks.

**Goals:**
* Explore the overall structure and characteristics of the dataset <br>
* Identify major patterns across industries and job titles <br>
* Examine trends related to job status, automation risk, growth, education, and salaries <br>
* Provide a foundation for the modeling and clustering notebooks 

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("ai_job_trends_dataset.csv")
data
```

### 1. Industries and Jobs


1.1 Count how many jobs belong to each industry and show which industries appear most often

```python
industry_count = data["Industry"].value_counts()
industry_count.plot(kind = "bar")
```

1.2 List all job titles and how many times each one appears in the dataset

```python
job_counts = data["Job Title"].value_counts().reset_index()

job_counts.columns = ["Job Title", "Count"]

job_counts
```

1.3 Inspect all rows for a specific job title to compare how entries for the same role differ

```python
dentist_table = data[data["Job Title"] == "Dentist"]
dentist_table
```

1.4 Find which job titles appear most often among the jobs that are decreasing

```python
decreasing_jobs = data[data["Job Status"] == "Decreasing"]
job_counts = decreasing_jobs["Job Title"].value_counts()
job_counts
```

1.5 How many decreasing jobs are there in each industry?

```python
decreasing_counts = decreasing_jobs["Industry"].value_counts()
decreasing_counts
```

1.6 Increasing vs Decreasing jobs by Industry

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.countplot(data=data, x="Industry", hue="Job Status")
plt.xticks(rotation=45)
plt.title("Increasing vs Decreasing Jobs by Industry")
plt.show()
```

### 2. Automation and Job Risk

2.1 Average automation risk per industry

```python
median_risk_industry = (
    data.groupby("Industry")["Automation Risk (%)"]
        .mean()
        .reset_index()
        .sort_values("Automation Risk (%)", ascending=False)
)

median_risk_industry
```

2.2 Average automation risk per job title

```python
median_risk_job = (
    data.groupby("Job Title")["Automation Risk (%)"]
        .mean()
        .reset_index()
        .sort_values("Automation Risk (%)", ascending=False)
)

median_risk_job
```

2.3 Job title with the highest automation risk within each industry

```python
risk_by_job_industry = (
    data.groupby(["Industry", "Job Title"])["Automation Risk (%)"]
        .median()
        .reset_index()
)

most_at_risk = (
    risk_by_job_industry.loc[
        risk_by_job_industry.groupby("Industry")["Automation Risk (%)"].idxmax()
    ]
    .sort_values("Automation Risk (%)", ascending=False)
)

most_at_risk
```

2.4 Automation Risk (%) vs Projected Growth. How many jobs fewer/more are expected in 2030 than there are in 2024?

```python
data["Projected Growth"] = data["Projected Openings (2030)"] - data["Job Openings (2024)"]

# Round Automation Risk to nearest 5% (0-4, 5-9 ...)
data["Risk Bin"] = (data["Automation Risk (%)"] // 5) * 5

summary = data.groupby("Risk Bin")["Projected Growth"].mean().reset_index()

plt.figure(figsize=(10,6))
plt.plot(summary["Risk Bin"], summary["Projected Growth"], marker="o")
plt.xlabel("Automation Risk (%)")
plt.ylabel("Average Projected Growth")
plt.title("Average Projected Growth by Automation Risk")
plt.grid(True)
plt.show()
```

### 3. Salaries and Education

3.1 Education distribution in each industry

```python
common_education = data.groupby("Industry")["Required Education"].agg(lambda x: x.mode()[0])

plt.figure(figsize=(10,6))
ax = sns.countplot(data=data, y="Industry", hue="Required Education",
              order=data["Industry"].value_counts().index)
for p in ax.patches:
    width = p.get_width()
    if width > 0:
        ax.text(width + 15, p.get_y() + p.get_height()/2, int(width), va="center", fontsize=7)
plt.xlim(0, 1100)
plt.title("Distribution of Education Levels per Industry")
plt.show()
```

3.2 Average salary by industry

```python
salary_by_industry = (data.groupby("Industry")["Median Salary (USD)"]
                      .mean()
                      .reset_index()
                      .sort_values("Median Salary (USD)", ascending=False))
print(salary_by_industry)
```

3.3 Average salary by job

```python
salary_by_job = (
    data.groupby("Job Title")["Median Salary (USD)"]
        .mean()
        .reset_index()
        .sort_values("Median Salary (USD)", ascending=False)
)

print(salary_by_job)
```

### 4. Countries and Geographic Insights

4.1 Average automation risk by country

```python
country_risk = data.groupby("Location")["Automation Risk (%)"].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
country_risk.plot(kind="bar", color="skyblue")
plt.xlabel("Country")
plt.ylabel("Average Automation Risk (%)")
plt.title("Average Automation Risk by Country")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

4.2 Average projected growth by country

```python
data["Projected Growth"] = data["Projected Openings (2030)"] - data["Job Openings (2024)"]
country_growth = data.groupby("Location")["Projected Growth"].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
country_growth.plot(kind="bar", color="lightgreen")
plt.xlabel("Country")
plt.ylabel("Average Projected Growth")
plt.title("Average Projected Growth by Country")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

4.3 Most common industries per country

```python
top_industries = data.groupby("Location")["Industry"].apply(lambda x: x.value_counts().idxmax())
top_industries
```

4.4 Most common jobs per country

```python
top_jobs = data.groupby("Location")["Job Title"].apply(lambda x: x.value_counts().idxmax())
top_jobs
```

4.6 Gender diversity by country

```python
country_gender = (data.groupby("Location")["Gender Diversity (%)"].mean().sort_values(ascending=False))

plt.figure(figsize=(12,6))
country_gender.plot(kind="bar", color="plum")
plt.xlabel("Country")
plt.ylabel("Average Gender Diversity (%)")
plt.title("Average Gender Diversity by Country")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

### 5. Unrealistic or Questionable findings:

#### 1. Industries and Jobs
* When examining the entries for dentist, we found dentists listed in industries such as finance, IT, manufacturing, and entertainment, which is highly unlikely.
* Two dentists were both placed in entertainment, but one required 0 years of experience and the other required 14 years, suggesting inconsistent or unreliable data.

#### 2. Automation and Job Risk
* Across almost all industries, the average automation risk was clustered tightly around 49–51%, which is suspiciously uniform.
* In the Automation Risk vs Projected Growth plot, jobs with higher automation risk did not show lower projected openings; instead, the points appeared almost random, indicating no meaningful correlation.

#### 3. Salaries and Education
* In the healthcare industry, the most common recorded education level was high school, which contradicts typical real-world requirements.
* Art gallery manager appeared among the highest-earning jobs, while veterinary surgeon ranked among the lowest - an outcome that does not match real-world salary expectations.

#### 4. Countries and Geographic Insights
* Average automation risk and gender diversity appear to cluster around 50% for nearly all countries, which is unlikely to reflect real-world differences.
